```python
def get_empty_candidate_stats():
    candidate_stats = {}
    for candidate in candidate_ref:
        candidate_stats[candidate] = { 'tweet_count': 0, 'pos': 0, 'neg': 0, 'neu': 0}
    return candidate_stats
```

```python
sentiments = ['pos', 'neg', 'neu']
def get_tweet_sentiment(tweet):
    blob = TextBlob(tweet)
    if abs(blob.sentiment[0]) < 0.3:
        return 'neu'
    if blob.sentiment[0] >= 0.3:
        return 'pos'
    if blob.sentiment[0] <= -0.3:
        return 'neg'
```

```python
def get_candidates_referred(tweet_text):
    refers_to = []
    for cand in candidate_ref:
        for ref in candidate_ref[cand]:
            if ref in tweet_text:
                refers_to.append(cand)
                break
    return refers_to
```

```python
def get_stats_for_tweet(tweet_text):
    # detect the sentiment of the tweet
    sentiment = get_tweet_sentiment(tweet_text)
    
    references_to = get_candidates_referred(tweet_text)
    
    return {'sentiment': sentiment, 'references_to': references_to}
```

```python
report = final_report['stats']
last_file_read = final_report['last_zip_read']
total_count = str(len(output_zip_paths))
try:
    count = 0
    seen_ids = set()
    i = 0
    for file_path in output_zip_paths:
        i += 1
        if count == 2:
            break
        try:
            # read all the contents of the file
            file_handler = gzip.open(file_path)
            contents = file_handler.read().decode("utf-8")
            file_handler.close()
            if contents == "":
                continue

            # mark a successful read
            count += 1

            #all_tweets.append(contents)
            # split the contents by '\n'. We now have an array of tweet strings
            this_tweets = contents.split('\n')

            # convert all the tweets to an object
            this_tweets_json = [json.loads(tweet) for tweet in this_tweets if tweet != ""]

            # iterate over all the tweet objects - Filter by date, find references to candidates, and find stats
            for tweet_json in this_tweets_json:
                filtered_tweet = {}

                # get the date of the tweet object
                tweet_date = datetime.strptime(tweet_json['created_at'], '%a %b %d %X %z %Y').strftime('%d/%m/%Y')

                # build a new tweet object with less variables
                filtered_tweet['created_at'] = tweet_json['created_at']
                filtered_tweet['id'] = tweet_json['id']
                filtered_tweet['text'] = tweet_json['text']
                
                if tweet_json['id'] not in seen_ids:
                    seen_ids.add(tweet_json['id'])
                else:
                    continue

                # detect who this tweet is referred to
                # get tweet stats returns the list of candidates the tweet has been referred to 
                # and the sentiment of the tweet
                tweet_stats = get_stats_for_tweet(tweet_json['text'])

                # if a tweet has not been seen on the tweet_date, then create a new object
                if tweet_date not in report:
                    report[tweet_date] = get_empty_candidate_stats()

                for cand in tweet_stats['references_to']:
                    report[tweet_date][cand][tweet_stats['sentiment']] += 1
                    report[tweet_date][cand]['tweet_count'] += 1
            
            last_file_read = file_path
            print("Read contents of file: "+file_path+" - "+ str(i) + "/" + total_count)
        except Exception as e:
            print("Exception while reading file: "+file_path + " - " + str(i) + "/"+total_count)
            print(format(e))
except KeyboardInterrupt:
    print("Interrupt detected!")
finally:
    print("Saving state...")
    final_report = {'last_zip_read': last_file_read, 'stats': report}
    content = json.dumps(final_report)
    f = open(os.path.join(REPORT_DIR, 'report.json'), "w")
    f.write(content)
    f.close()
    print("State saved")
```

## Graphing the data
We read the report from the disk.

```python
report_file = open(os.path.join(REPORT_DIR, 'report.json'), "r")
report_contents = report_file.read()
report_file.close()
final_report = json.loads(report_contents)
```

```python
def build_graphable_data_from_report(report):
    dates = sorted(report.keys(), key=lambda x: datetime.strptime(x, '%d/%m/%Y'))
    candidate_time_series = {}
    for cand in candidate_ref:
        candidate_time_series[cand] = {'tweet_count': [], 'pos': [], 'neg': [], 'neu': []}

    for date in dates:
        for candidate in candidate_ref:
            candidate_time_series[candidate]['tweet_count'].append(report[date][candidate]['tweet_count'])
            candidate_time_series[candidate]['pos'].append(report[date][candidate]['pos'])
            candidate_time_series[candidate]['neg'].append(report[date][candidate]['neg'])
            candidate_time_series[candidate]['neu'].append(report[date][candidate]['neu'])
    return (dates, candidate_time_series)
```

```python
(dates, candidate_time_series) = build_graphable_data_from_report(final_report['stats'])
candidate_time_series
```

```python
fig, ax = plt.subplots(12, 1, sharey='row', figsize = (20,60), sharex= False)
i = 0
for cand in candidate_ref:
    #plt.subplot(12, 1, i)
    ax[i].plot(dates, candidate_time_series[cand]['tweet_count'], label = 'Tweet count')
    ax[i].plot(dates, candidate_time_series[cand]['pos'], label = "Positive")
    ax[i].plot(dates, candidate_time_series[cand]['neg'], label = "Negative")
    ax[i].plot(dates, candidate_time_series[cand]['neu'], label = "Neutral")
    ax[i].title.set_text(cand)
    ax[i].legend()
    i+=1 
```
