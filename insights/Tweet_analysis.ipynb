```python
import numpy as np
import pandas as pd
import json
from simplejson import JSONDecodeError
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
file_1 = open("/data/us2020/sample1/output_1.txt")
output_1_data = file_1.read()
file_1.close()
tweets = []
```

```python
lines = output_1_data.split('\n')
```

```python
tweets = []
for line in lines:
    try:
        tweet = json.loads(line)
        tweets.append(tweet)
    except:
        print('Problems for - \n')
        print(line)
```

```python
pd_tweets = pd.DataFrame(tweets)
```

```python
simpletime = [datetime.strptime(tweet['created_at'], '%a %b %d %X %z %Y').strftime('%d/%m/%Y') for tweet in tweets]
```

```python
pd_tweets['simpletime'] = simpletime
```

```python
candidate_ref = {'Elizabeth Warren': ['@ewarren', '#imWithHer', '#TeamWarren', '#Warren', '#Warren2020', '#WarrenHarris', '#WarrenIsaSnake', '#WarrenCastro2020'],
                 'Michael Bennet':['@MichaelBennet'],
                 'Joe Biden' : ['@JoeBiden', '#Biden2020', '#JoeBiden', '#TeamJoe', '#TellTheTruthJoe', '#BidenIsCorrupt'],
                 'Michael Bloomberg' : ['@Mike2020', '#bloomberg2020'],
                 'Pete Buttigieg' : ['@PeteButtigieg', '#dropoutpete', '#pete2020', '#PeteForAmerica','#PeteforPresident', '#TeamPete', '#winecavepete'],
                 'John Delaney' : ['@JohnDelaney'],
                 'Tulsi Gabbard' : ['@tulsigabbard', '#StandWithTulsi', '#tulsiwasright', '#tulsigabbard', '#Tulsi2020', '#Tulsi', '#tulsicoward'],
                 'Amy Klobuchar' : ['@amyklobuchar', '#AmyforAmerica', '#klobucharmy', '#DropOutAmy'],
                 'Deval Patrick' : ['@DevalPatrick', '#Devalforall', '#DevalPatrick2020', '#devalpatrick'],
                 'Bernie Sanders' : ['@BernieSanders', '#Bernie', '#berniesanders', '#bernie2020', '#Fightfor15', '#ILikeBernie', '#NobodyLikesHim', '#NotMeUs', '#TrustBernie', '#ITrustBernie', '#VoteforBernie', '#WomenForBernie', '#blackwomenforbernie'],
                 'Tom Steyer' : ['@TomSteyer', '#momenTOM', '#TomforPresident'],
                 'Andrew Yang': ['@AndrewYang', '#AmericaNeedsYang', '#AndrewYang', '#NHforYang', '#Yang2020', '#YangGang'],
                }
```

```python
ref_count = {}
tweets_by_candidate = {}
for candidate in candidate_ref:
    ref_count[candidate] = 0
    tweets_by_candidate[candidate] = []
    
# count the number of times each of the candidates are referred to in all the tweets
for index, tweet in pd_tweets.iterrows():
    for candidate in candidate_ref:
        refs = candidate_ref[candidate]
        for ref in refs:
            if ref in tweet['text']:
                ref_count[candidate] = ref_count[candidate] + 1
                tweets_by_candidate[candidate].append(tweet)
```

```python
tweet_cnt_by_date = {}
pd_tweetes_by_candidate = {}
for candidate in tweets_by_candidate:
    pd_tweetes_by_candidate[candidate] = pd.DataFrame(tweets_by_candidate[candidate])
    tweet_cnt_by_date[candidate] = pd_tweetes_by_candidate[candidate].groupby('simpletime').count()['created_at']
```

```python
tweet_cnt_by_date
```

```python
ref_count_ser = pd.Series(ref_count).sort_values()
objects = ref_count_ser.index
y_pos = np.arange(len(objects))
performance = ref_count_ser.values

plt.barh(y_pos, performance, align='center', alpha=0.5)
plt.yticks(y_pos, objects)
plt.xlabel('Tweets')
plt.title('Total tweets')

plt.show()
```

```python
for candidate in tweet_cnt_by_date:
    tweet_cnt_by_date[candidate].plot(legend = True, label = candidate, figsize = (18, 7))
plot = plt.xlabel('Time')
plot = plt.ylabel('Count')
```
